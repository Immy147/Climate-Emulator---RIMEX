```python
import xarray as xr
import matplotlib.pyplot as plt

ds = xr.open_dataset(
    "../Canesm5/pr/hist/pr_station.nc",
    decode_times=True,
    use_cftime=True
)

pr = ds['pr'].squeeze()
pr_mm_day = pr * 86400

# Annual sum
annual_pr = pr_mm_day.resample(time="Y").sum()

# Convert to pandas for plotting
annual_pr_pd = annual_pr.to_pandas()
annual_pr_pd.index = annual_pr_pd.index.year

annual_pr_pd.plot(figsize=(10,4))
plt.title("Annual Precipitation - CanESM5 Historical")
plt.ylabel("mm/year")
plt.grid()
plt.show()
```
